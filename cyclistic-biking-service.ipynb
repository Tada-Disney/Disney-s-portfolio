## Case Study: How does a Bike-Share Navigate Speedy Success

## Background
In 2016, Cyclistic launched a successful bike-share offering. Since then, the program has grown to a fleet of 5,824 bicycles that are geotracked and locked into a network of 692 stations across Chicago.The bikes can be unlocked from one station and returned to any other station in the system anytime.
Until now, Cyclistic’s marketing strategy relied on building general awareness and appealing to broad consumer segments. One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes, and annual memberships. Customers who purchase single-ride or full-day passes are referred to as casual riders. Customers who purchase annual memberships are Cyclistic members. 

## Ask
Three questions will guide the future marketing program:
1. How do annual members and casual riders use Cyclistic bikes differently?
2. Why would casual riders buy Cyclistic annual memberships?
3. How can Cyclistic use digital media to influence casual riders to become members?

### **DATA CLEANSING AND TRANSFORMATION**

#### The dataset is made up of 12 tables with the same number of columns and of the same type. Merging data to consolidate analysis 

```r
all_trips <- bind_rows(Jul2022, Aug2022, Sep2022, Oct2022, Nov2022, Dec2022, Jan2023, Fev2023, Mar2023, Apr2023, May2023, Jun2023)
```

#### A look on the newly created table(all_trips)

```r
head(all_trips)
```

```r
str(all_trips)
```

#### Getting rid of columns that won't be use for analysis 

```r
colnames(all_trips)
```

```r
all_trips <- all_trips%>%
select(-c(start_lat, end_lat, start_lng, end_lng, start_station_id, end_station_id, end_station_name))
```

#### A look on all_trips columns 

```r
colnames(all_trips)
```

#### The data is not in the right format and new column has to be created for date and time 

```r
all_trips$date <- as.Date(all_trips$started_at) ## the default format is yyyy-mm-dd
all_trips$month <- format(as.Date(all_trips$date), "%m")
all_trips$day <- format(as.Date(all_trips$date), "%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")
```

#### Let's calculate a field for each unique ride

```r
all_trips$ride_length <- difftime(all_trips$ended_at, all_trips$started_at)
```

#### let's have a look on all_trips structure

```r
str(all_trips)
```

#### Let's convert all-trips ride-length from factor to numeric value 

```r
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))
is.numeric(all_trips$ride_length)
```

#### Let's remove blancs

```r
all_trips <- all_trips[!(all_trips$start_station_name == "HQ QR"|all_trips$ride_length<0),]
```

#### Setting days of the week to the order Sunday to Saturday

```r
all_trips$day_of_week <- ordered(all_trips$day_of_week, level=c("sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"))
```

#### Let's creat a column for weekday field

```r
all_trips %>%
  mutate(weekday = wday(started_at, label = TRUE)) %>%
  group_by(member_casual, weekday) %>%
  summarise(number_of_ride = n(), .groups = "keep")
```

### **DATA VISUALISATION**

```r
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")

result <- all_trips %>%
  group_by(member_casual, day_of_week) %>%
  summarise(number_of_ride = n(), .groups = "keep") %>%
  arrange(member_casual, day_of_week)

library(ggplot2)
ggplot(result, aes(x = day_of_week, y = number_of_ride, fill = member_casual)) +
  geom_col(position = "dodge") +
  labs(title = 'Ride per Day of Week')
```

- From the graph, the data shows that casual riders tends to use bikes more frequently on weekends, while members tend to use them more frequently on weekdays.

```r
all_trips %>%
  mutate(weekday = wday(started_at, label = TRUE)) %>%
  group_by(member_casual, weekday) %>%
  summarise(number_of_rides = n(), average_duration = mean(ride_length), .groups = "keep") %>%
  arrange(member_casual, weekday) %>%
  ggplot(aes(x = weekday, y = average_duration, color = member_casual)) +
  geom_line(aes(group = member_casual)) +
  labs(title = 'Average Ride Time per Week')
```

- The data shows that casual riders overall spend more time using the service than their full-time counterparts.

```r
all_trips%>%
group_by(member_casual, month)%>%
summarise(total_ride = n(), 'average_duration_(minS)'= mean(ride_length), .groups = 'keep')%>%
arrange(member_casual)%>%
ggplot(aes(x= month, y= total_ride, fill = member_casual))+ geom_col(position = "dodge")+
labs(title= 'Total Ride per Month')
```

- The data shows that casual users were more attive during the summer month than their long-term, while the winter monther shows little sign of activity of casual users. 

```r
all_trips%>%
ggplot(aes(x= rideable_type, fill = member_casual))+ geom_bar(position = "dodge") + 
labs(x= 'type of bike', y='number of rentals', title= 'Which Bike Work the Most')
```

- The data shows both types of memberships prefer using the classic bike more than the electric bike. Adiitionally long-term memebers are also seen to favour classic bike over the electric bike.

### **INSIGHTS**
- The data shows that casual riders tend to use the bikes more frequently on weekends, while members tend to use them more frequently on weekdays. This suggests that members are more likely to use the bikes for commuting purposes rather than for leisure activities
- The data shows that casual riders were more active during the summer months than the long-term, while the winter months show very little activity on the part of the casual users. In contrast, long-term users are more active in the winter and spring months.
- The data shows that both types of memberships prefer using the classic bike more than the electric bike. Additionally, long-term members are also seen to favor the classic bike over the electric bike
